```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('economy.csv')
df.head()
```

```python
df.info()
```

```python
df['price'] = df['price'].astype(str).str.replace(',', '', regex=False)
df['price'] = df['price'].astype(int)
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Date']= df['date'].str.split('-').str[0]
df['Month']= df['date'].str.split('-').str[1]
df['Year']= df['date'].str.split('-').str[2]
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month']= df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
#Drop Date column
df.drop('date', axis=1, inplace = True)
```

```python
df.head()
```

```python
df['Arrival_Hour'] = df['arr_time'].str.split(':').str[0]
df['Arrival_Minute'] = df['arr_time'].str.split(':').str[1]
```

```python
df.info()
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
df.info()
```

```python
df.drop('arr_time', axis=1, inplace = True)
df.head()
```

```python
df['Departure_Hour'] = df['dep_time'].str.split(':').str[0]
df['Departure_Minute'] = df['dep_time'].str.split(':').str[1]
df.info()
```

```python
df['Departure_Hour'] = df['Departure_Hour'].astype(int)
df['Departure_Minute'] = df['Departure_Minute'].astype(int)
df.info()
```

```python
df.drop('dep_time', axis=1, inplace = True)
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('ch_code', axis=1, inplace=True)
df.head()
```

```python
df.head()
```

```python
df['Total_hours'] = df['time_taken'].str.extract(r'(\d+)h').fillna(0).astype(int)
df['Total_minutes'] = df['time_taken'].str.extract(r'(\d+)m').fillna(0).astype(int)
df.head()
```

```python
df.drop('time_taken', axis = 1, inplace = True)
df.head()
```
